```python
import pandas as pd
import numpy as np
import json
from datetime import datetime

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df.state.value_counts()
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,5))
labels = ['successful', 'failed']
colors = ['cyan', 'tomato']
explode = (0, 0.1)
ax[0].pie(df.state.value_counts(), startangle=90, colors=colors, wedgeprops={'edgecolor': 'black'}, autopct='%1.f%%',
        explode=explode, shadow=True,labels=labels,)
ax[0].set_title('Kickstater State Distribution - Pie Chart')

sns.countplot(x=df.state,ax=ax[1],palette=colors)
for idx, row in pd.DataFrame(df['state'].value_counts().reset_index()).iterrows():
    ax[1].text(idx,row.state, str(round(row.state)),color='black', ha="center")
ax[1].set_title("Count Plot")
plt.tight_layout()
plt.show()
```

```python
def fxn_hist_box_plot(df,col):
    fig,ax=plt.subplots(1,2,figsize=(12,5))
    #sns.violinplot(df[col],ax=ax[0],color='teal',cut=2)
    sns.boxplot(df[col],ax=ax[0],color='teal')
    ax[0].set_title('Violin plot')
    
    ax[0].set_xlabel(col)


    sns.distplot(df[col],ax=ax[1],color='tomato')
    ax[1].axvline(df[col].mean(),  label='Mean',color='teal')
    ax[1].axvline(df[col].median(),  label='Median',linestyle='dotted',color='black')
    ax[1].set_xlabel(col)

    ax[1].set_title('Histogram')
    tit1='Distribution of '+col
    plt.suptitle(tit1)
    plt.tight_layout()
    plt.legend()
    plt.show()
```

```python
fxn_hist_box_plot(d1,'daysToDeadline')
```
